```python
#Elimino las tildesy y los valores ascii que aparecian C¢rdoba' 'Entre R¡os' 'Neuqu\x82n' 'R¡o Negro'
def limpiar_texto(x):
    if pd.isna(x):
        return x
    # Normaliza y elimina tildes y caracteres raros
    x = unicodedata.normalize("NFKD", x).encode("ascii", "ignore").decode("utf-8")
    return x

# Limpieza + Title Case sin tildes
df['jurisdiccion'] = df['jurisdiccion'].astype(str).apply(limpiar_texto).str.title()
#Obtener los valores unicos de la columa "jurisdiccion" ya capitalizados
cavoles_unicos = df['jurisdiccion'].unique()
#Imprimir los valores unicos
print("Valores unicos de la columna 'jurisdiccion':")
print(cavoles_unicos)
```

```python
#ACa observo que 'sexo' , que hay valores 'ambos_sexos' 'mujeres' 'varones' 'Ambos sexos' 'Mujeres' 'Varones']
#reemplazar los valores  que qeudaron medio mal para que se entiendan mejor
#replazamos ambos_sexos' 'mujeres' 'varones' por 'Ambos sexos' 'Mujeres' 'Varones' en la columna 'sexo'

df['sexo']=df['sexo'].replace({'ambos_sexos' : 'Ambos sexos', 'mujeres' : 'Mujeres', 'varones' : 'Varones'})

#Obtener loos calores unicos actualizados para la columna 'sexo'
valores_unicos_sexo = df['sexo'].unique()

#Imprimir los valores unicos actualizados
print("Valores unicos actualizados de la columna 'sexo':")
print(valores_unicos_sexo)
```

```python
#Observo que los valores de la tasa de sifilis, agunos valores estas con " " y otrso con . o con , . Esto me indica que hay diferentes tipos de datos dando vueltas, string, decimales y float , lo mas conveniente es pasar todo a float ya que el valor es unaTASA y lleva comas ;D

df['jurisdiccion_tasa_sifilis'] = (
    df['jurisdiccion_tasa_sifilis']
    .astype(str)                        # pasar todo a string
    .str.strip()                        # eliminar espacios
    .str.replace("'", "", regex=False)  # eliminar comillas simples
    .str.replace('"', "", regex=False)  # eliminar comillas dobles
    .str.replace(",", ".", regex=False) # coma → punto
)

df['jurisdiccion_tasa_sifilis'] = pd.to_numeric(df['jurisdiccion_tasa_sifilis'], errors='coerce')
```
